```python
from kokoro import KPipeline
import sounddevice as sd # Requires: pip install sounddevice

# Initialize Pipeline (lang_code 'a' for American English)
pipeline = KPipeline(lang_code='a')

text = "Kokoro is an open-weight TTS model that is fast enough for real-time streaming."

# Generator to stream chunks
generator = pipeline(text, voice='af_heart')

# Stream audio chunks
for i, (gs, ps, audio) in enumerate(generator):
    print(f"Playing chunk {i}")
    # Play audio immediately
    sd.play(audio, 24000)
    sd.wait()
```
